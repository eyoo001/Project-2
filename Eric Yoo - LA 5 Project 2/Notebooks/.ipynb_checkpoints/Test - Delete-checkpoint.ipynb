```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.linear_model import Lasso, LassoCV, LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, VarianceThreshold, f_regression, SelectPercentile
from sklearn.grid_search import GridSearchCV
from sklearn.feature_selection import variance_threshold
np.random.seed(42)


%matplotlib inline
```

```python
df = pd.read_csv('../Data/train.csv', index_col='Id')
```

```python
df.drop(['PID'], axis=1, inplace=True)
```

```python
df_obj = df.select_dtypes(include = ['object'])
df_int = df.select_dtypes(exclude = ['object'])
```

```python
df_obj_d = pd.get_dummies(df_obj, drop_first=True)
```

```python
df_con = pd.concat((df_int, df_obj_d), axis=1)
df_con.fillna(0, inplace=True)
```

```python
df_con.shape
```

```python
X = df_con.drop(['SalePrice'], axis=1)
y = df['SalePrice'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
pipe = Pipeline([
    ('var_thresh', VarianceThreshold(.05)),
    ('ss', StandardScaler()),
    ('kbest', SelectKBest(f_regression, k=5)),
    ('lr', LinearRegression())
])
```

```python
params = {
    'var_thresh__threshold': [0, .05, .15],
    'kbest__k': [9, 35, 45]
#    'kpercentile__percentile': (10, 25)
}
```

```python
gs = GridSearchCV(pipe, params)
gs.fit(X_train, y_train)
```

```python
gs.score(X_train, y_train)
```

```python
gs.score(X_test, y_test)
```

```python
gs.best_params_
```

```python
gs.best_estimator_.named_steps['lr'].intercept_
```

```python
betas = pd.DataFrame(
    gs.best_estimator_.named_steps['lr'].coef_,          
    index = X.columns[gs.best_estimator_.named_steps['var_thresh'].get_support()]
    [gs.best_estimator_.named_steps['kbest'].get_support()], 
#    [gs.best_estimator_.named_steps['kpercentile'].get_support()], 
    columns=['weight'])
```

```python
betas['abs_w'] = betas.weight.abs()

weights = betas.sort_values('abs_w', ascending=False)['weight'].head(8)

labels = weights.index

beta_0 = gs.best_estimator_.named_steps['lr'].intercept_
```

```python
plt.figure(figsize=(20,10))
plt.barh(labels, weights)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Weights', fontsize=30)
plt.title(f'Most Important Features and Weights\ny intercept: ${beta_0:0.02f}', fontsize=42)
plt.tight_layout()
```
